```python
import pandas as pd
import statistics
from datetime import date
```

```python
#import R and S spreadsheets
rly = pd.read_excel('/content/R-Copy.xlsx', index_col=0)
st = pd.read_excel('/content/S-Copy.xlsx',sheet_name=1)
```

```python
#set start and end date of report
startday = date(2021,1,1)
endday = date(2021,12,12)
```

```python
def diff_dates(date1, date2):
    return abs(date2-date1).days

d0 = date(2017,9,1)
d1 = date(2014,6,25)

d2 = startday
d3 = endday
sk0 = diff_dates(d2, d1)-1
sk1 = diff_dates(d3, d1)
rk0 = diff_dates(d2, d0)-1
rk1 = diff_dates(d3, d0)

print(sk0)
print(sk1)
print(rk0)
print(rk1)
```

```python
st0 = st.iloc[2:38, sk0:sk1]
#st0 = st.iloc[2:38, 9:2725]
st1 = st0.T

#2725 = 12/10/21
```

```python
rly0 = rly.iloc[3:61, rk0:rk1]
#rly0 = rly.iloc[3:61, 4:1561]
rly1 = rly0.T

#1561 = 12/10/21
```

```python
st1.to_csv('S_out.csv', index=False)
rly1.to_csv('R_out.csv', index=False)
```
